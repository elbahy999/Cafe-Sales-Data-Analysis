```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('dirty_cafe_sales.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Item'].value_counts()
```

```python
df = df.replace(['UNKNOWN', 'ERROR'], np.nan)
```

```python
numeric_cols = ["Price Per Unit", "Quantity", "Total Spent"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Item'].value_counts()
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Total Spent'] / df['Price Per Unit'])

df = df.dropna(subset=['Quantity'])
```

```python
df['Quantity'].value_counts()
```

```python
df['Item'].value_counts()
```

```python
df.isnull().sum()
```

```python
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])

df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Total Spent'] / df['Quantity'])

```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Price Per Unit'])
df = df.dropna(subset=['Total Spent'])

df = df.reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python

price_item_counts = (
    df
    .groupby(["Price Per Unit", "Item"])
    .size()
    .reset_index(name="Count")
    .sort_values(["Price Per Unit", "Count"], ascending=[True, False])
)
print(price_item_counts)
```

```python
price_to_item = {
    1.50: "Tea",
    2.00: "Coffee",
    5.00: "Salad",
    1.00: "Cookie"
}

def fix_item(row):
    if pd.isna(row["Item"]):
        price = row["Price Per Unit"]
        
        if price == 3.00:
            return random.choice(["Cake", "Juice"])
        elif price == 4.00:
            return random.choice(["Sandwich", "Smoothie"])
        
        return price_to_item.get(price, row["Item"])

    return row["Item"]

df["Item"] = df.apply(fix_item, axis=1)
```

```python
df.isnull().sum()
```

```python
df['Item'] = df['Item'].str.strip().str.capitalize()
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
df = df.dropna(subset=['Transaction Date'])

df['Payment Method'] = df['Payment Method'].fillna('Unknown')
df['Location'] = df['Location'].fillna('Unknown')

df = df.reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Day_of_Week'] = df['Transaction Date'].dt.dayofweek
df['Month'] = df['Transaction Date'].dt.month
df['Day'] = df['Transaction Date'].dt.day
df['Week_of_Year'] = df['Transaction Date'].dt.isocalendar().week
df['Is_Weekend'] = (df['Day_of_Week'] >= 5).astype(int)
df['Quarter'] = df['Transaction Date'].dt.quarter

df['Day_Name'] = df['Transaction Date'].dt.day_name()

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['Month'].apply(get_season)

df.reset_index(drop=True, inplace=True)

df.to_csv('cleaned_cafe_sales.csv', index=False)
```
